## Visualization:

### Correlation Heat Map:

```python
correlation_matrix = df_tracks.corr()
plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt = "0.1g", vmin = -1, vmax = 1, center = 0, cmap="coolwarm")
heatmap.set_title("Correlation Heatmap between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90);
```

### Creating Regression Plots:

```python
sample_df = df_tracks.sample(int(0.004 * len(df_tracks)))  # Taking a fraction of a data as a sample.
```

```python
len(sample_df)
```

#### Creaing a regression plot for Loudness and Energy:

```python
plt.figure(figsize = (8,6))
sns.regplot(data = sample_df, y = "loudness", x = "energy", color = 'darkgreen').set(title = "Loudness VS. Energy Correlation");
```

#### Analysis:
* Upward regression line.
* There is a highly positive correlation between loudness and energy.
* The louder tracks have higher energy and vice versa. 

#### Creating regression plot for popularity and acousticness:

```python
plt.figure(figsize= (8,6))
sns.regplot(data = sample_df, y = "popularity", x = "acousticness", color = "teal").set(title = "Popularity Vs. Acousticness");
```

#### Analysis:
* Downward regression line.
* There is a negetive correlation between popularity and acousticness.
* The acousticness of the tracks increase, the popularity decrease.

#### Creating a distribution plot to visualize the total number of songs in each year since 1922 available in the Spotify streaming:

```python
years.plot(kind = 'hist', figsize = (10,6), bins = 20, color ='skyblue', ec = 'darkblue')
plt.title("Number of songs per year")
plt.xlabel("Year")
plt.ylabel("Number of Songs");
```

#### Analysis:
* The plot tells us that the songs released in recent years have increased as music became accessable to people globally with technological advancement.

### Duration of Songs over the years:

#### Plotting a bar graph:

```python
fig, ax = plt.subplots(figsize=(15, 7))
sns.barplot(x=years, y=df_tracks['duration_sec'], ax=ax, errwidth=False, ec= 'black')

plt.title("Year vs Duration")
plt.xticks(rotation=90);
```

#### Plotting a line graph:

```python
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(10, 5))
sns.lineplot(x=years, y=df_tracks['duration_sec'], ax=ax)

plt.title("Year vs Duration")
plt.xticks(rotation=60);
```

#### Analysis:
* In the x-axis, we have year of release and at the y_axis, we have duration in seconds. Data is from 1922 till 2021.
* We can see that initially, the songs during early 1900s to mid 1900s, the songs had shorter duration. This may be due to the tracks have more vocals than instruments based.
* As the years went by, the tracks had more duration as both vocals and instrument played a signuficant role in the composition of sound tracks.
* As years went by, from 2010 onwards the duration of the sound tracks declined as the attention span of the average listners declined.
